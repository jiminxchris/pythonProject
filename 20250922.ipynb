```python
# 도전미션
리스트 numbers 각 원소는 숫자 0부터 9까지로 이루어져 있습니다. 이때, numbers에서 연속적으로 나타나는 숫자는 하나만 남기고 전부 제거하려고 합니다. 단, 제거된 후 남은 수들은 numbers의 원소의 순서를 유지해야 합니다. 
예를 들면,
numbers = [1, 1, 3, 3, 0, 1, 1] ->  [1, 3, 0, 1] 
numbers = [4, 4, 4, 3, 3] ->  [4, 3]
코드를 작성해 주세요
```

```python
# numbers = [1, 1, 3, 3, 0, 1, 1]
numbers = [4, 4, 4, 3, 3]

answer = []
comp = numbers[0]
answer.append(comp)

for i in numbers:
    if comp != i: # 이전숫자와 같은지 비교
        comp = i # 다르다면 비교할 숫자 갱신
        answer.append(comp) # 새로운 숫자이므로 리스트에 삽입

answer
```

```python
def max_num(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val
```

```python
num = max_num( -3, -5, -1)
num # 12
```

```python
def num_min(arr):
    answer = arr[0]
    for i in arr:
        if answer > i:
            answer = i
    return answer
```

```python
num_min([52, 273, 32 ,103, 275, 24, 57]) # 24
```

```python
num_min([-3, -5, -1]) # -5
```

```python
def morse_to_string(letter):
    morse = {
    '.-':'a','-...':'b','-.-.':'c','-..':'d','.':'e','..-.':'f',
    '--.':'g','....':'h','..':'i','.---':'j','-.-':'k','.-..':'l',
    '--':'m','-.':'n','---':'o','.--.':'p','--.-':'q','.-.':'r',
    '...':'s','-':'t','..-':'u','...-':'v','.--':'w','-..-':'x',
    '-.--':'y','--..':'z'
    }
    # 코드구현
    a = letter.split()
    answer = ''
    for i in a:
        answer += morse[i]
    return answer
```

```python
letter = ".... . .-.. .-.. ---"

morse_to_string(letter)
```

```python
def add_mul(choice, *args):
    # print(choice, args)
    # 여기에 코드를 작성해주세요!!!
    if choice == 'add':
        result = 0
        for i in args:
            result += i
        return result
    elif choice == 'mul':
        result = 1
        for i in args:
            result *= i
        return result
    else:
        print('연산자 종류 확인 요망!!!')
        return None
```

```python
add_mul('add', 1, 2, 3, 4, 5)
```

```python
add_mul('mul', 1, 2, 3)
```
